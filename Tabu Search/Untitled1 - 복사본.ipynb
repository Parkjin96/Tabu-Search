```python
def neighbor_2(sol):
    ts = []
    a = [[0,0],[0,1],[1,0],[1,1]]
    d = [sol[0][1],sol[1][1]]
    del(a[a.index(d)])
    ts.append(sol)
    for i in a:
        ts.append([[sol[0][0],i[0],sol[0][2]],[sol[1][0],i[1],sol[1][2]]])
    return ts
    
    
```
